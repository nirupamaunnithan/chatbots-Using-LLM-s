```python
from transformers import AutoTokenizer, AutoModelForCausalLM

tokenizer = AutoTokenizer.from_pretrained("cognitivecomputations/dolphin-2.2-mistral-7b")
model = AutoModelForCausalLM.from_pretrained("cognitivecomputations/dolphin-2.2-mistral-7b")
```

```python
instruction = (
    "You are an anger management assistant and a supportive friend. "
    "Your role is to help users feel calm and provide them with empathy and understanding. "
    "Respond in a gentle and encouraging way, offering advice to help them manage their emotions."
)
```

```python
def generate_response(user_input):
    prompt = f"{instruction}\n\nUser: {user_input}\nAssistant:"
    inputs = tokenizer(prompt, return_tensors="pt")

    # Generate a response with the model
    outputs = model.generate(inputs.input_ids, max_length=150, num_return_sequences=1, pad_token_id=tokenizer.eos_token_id)
    response = tokenizer.decode(outputs[0], skip_special_tokens=True)

    return response

# Example usage
user_input = "I am angry at my neighborhood aunty because she said why i dont have a boyfriend"
response = generate_response(user_input)
print(response)
```
